## Question 3a

Completions between years: What are the percent of different degrees completed for a given year range and sex? Parameter arguments are as follows: two year arguments, and a value for sex (A, F, or M). Function should return all rows of the data which match the given sex, and have data between the given years (inclusive for the start, exclusive for the end). If no data is found for the parameters,return Python keyword None.

```python
def completions_between_years(start_year,end_year,gender):
    return(data.loc[(data['Sex'] == gender) &
              (data['Year']>=start_year) & (data['Year']<end_year)])
```

## Question 3b

Compare bachelors in year: What were the percentages for women vs men having earned a Bachelors Degree in a given year? Parameter list argument is the year in question and return the percentages as a tuple: (% for men, % for women).

```python
def compare_bachelors_in_year(year):
    a=data.loc[(data['Min_degree'] == "bachelor's") &(data['Year']==year) & (data['Sex']!='A')][['Sex','Total']]
    records = a.to_records(index=False)
    result = list(records)
    return(result)
```

```python
while True:
  try:
    year=int(input('Enter a year from 1920 to 2018:'))
    # Check if input is in range
    if year in range(1920,2019):
      break
    else:
      print ('Your number is out of the range of 1920 to 2018')
  except:
    print ("You have not entered a number. Please try again")
```

```python
#returns tuple in the following format (% for men, % for women)
compare_bachelors_in_year(year)
```

## Question 3c

top 2 2000s: What were the two most commonly awarded levels of educational attainment awarded between 2000-2010 (inclusive)? Use the mean percent over the years to compare the education levels. Return a list of tuples as follows: [(#1level, mean % of #1 level), (#2 level, mean % of #2 level)].

```python
def top_2_2000s():
    new=data.loc[(data['Year']>=2000) & (data['Year']<=2010)]
    new1=new.iloc[:,2:4]
    new1['Min_degree'].value_counts()
    grouped_df = new1.groupby("Min_degree")
    mean_df = grouped_df.mean()
    mean_df=mean_df.sort_values(by='Total', ascending=False).iloc[0:2,]
    mean_df = mean_df.reset_index()
    records = mean_df.to_records(index=False)
    result = list(records)
    return(result)
```

```python
top_2_2000s()
```
